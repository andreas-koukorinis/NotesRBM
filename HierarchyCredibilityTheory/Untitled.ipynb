```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy.optimize import minimize
import statsmodels as sts
import seaborn as sns
```

```python
# Parameters
n_groups = 10              # e.g., regions
n_contracts_per_group = 20 # e.g., policyholders
mu = 100                   # global mean
tau2 = 25                  # group-level variance
sigma2 = 100               # individual-level variance

# Simulate data
group_effects = np.random.normal(loc=mu, scale=np.sqrt(tau2), size=n_groups)
data = []

for i in range(n_groups):
    group_mean = group_effects[i]
    contracts = np.random.normal(loc=group_mean, scale=np.sqrt(sigma2), size=n_contracts_per_group)
    for j, value in enumerate(contracts):
        data.append({'group': i, 'contract': j, 'loss': value})

df = pd.DataFrame(data)
df.head()


```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='group', y='loss')
plt.title("Loss Distribution by Group")
plt.show()
```
